```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("Fraud_Prediction.csv")
```

```python
dataset
```

```python
# Checking for missing values

dataset.isnull().sum()
```

```python
# Dropping columns which are not required for the analysis
dataset=dataset.drop(['Transaction_ID','User_ID'],axis=1)
dataset
```

```python
#Fetching month name from Timestamp

dataset['Timestamp']=pd.to_datetime(dataset['Timestamp'])
```

```python
dataset['Transaction_Month']=dataset['Timestamp'].dt.month_name()
```

```python
# Converting Card age in months to card age in years

dataset['Card_Age_Years'] = pd.cut(dataset['Card_Age'], bins=[0, 12, 60, 120, 250], labels=['0 to 1', '1 to 5', '5 to 10', '10+'])
```

```python
# Dropping Timestamp & Card age in months columns

dataset=dataset.drop(['Timestamp','Card_Age'],axis=1)
```

```python
# Checking for duplicate records

dataset.duplicated().sum()
```

```python
dataset.info()
```

```python
# Splitting Quantitative and Qualitative columns

quan=[]
qual=[]
def quanqual(dataset):
    for columnname in dataset.columns:
        if(dataset[columnname].dtype=='O'):
            qual.append(columnname)
        elif(dataset[columnname].dtype=='int64'):
            quan.append(columnname)
        elif(dataset[columnname].dtype=='float64'):
            quan.append(columnname)
    return quan,qual
```

```python
quanqual(dataset)
```

```python
descriptive=pd.DataFrame(index=['Q1-25%','Q2-50%','Q3-75%','99%','Q4-100%','IQR','1.5 RULE','Lesser','Greater','Min','Max'],columns=quan)
```

```python
for ColumnName in quan:
    descriptive[ColumnName]['Q1-25%']=dataset.describe()[ColumnName]['25%']
    descriptive[ColumnName]['Q2-50%']=dataset.describe()[ColumnName]['50%']
    descriptive[ColumnName]['Q3-75%']=dataset.describe()[ColumnName]['75%']
    descriptive[ColumnName]['99%']=np.percentile(dataset[ColumnName],99)
    descriptive[ColumnName]['Q4-100%']=dataset.describe()[ColumnName]['max']
    descriptive[ColumnName]['IQR']=descriptive[ColumnName]['Q3-75%']-descriptive[ColumnName]['Q1-25%']
    descriptive[ColumnName]['1.5 RULE']=1.5*descriptive[ColumnName]['IQR']
    descriptive[ColumnName]['Lesser']=descriptive[ColumnName]['Q1-25%']-descriptive[ColumnName]['1.5 RULE']
    descriptive[ColumnName]['Greater']=descriptive[ColumnName]['Q3-75%']+descriptive[ColumnName]['1.5 RULE']
    descriptive[ColumnName]['Min']=dataset[ColumnName].min()
    descriptive[ColumnName]['Max']=dataset[ColumnName].max()
```

```python
descriptive
```

```python
# Finding Outliers in the data


lesser=[]
greater=[]
for ColumnName in quan:
    if(descriptive[ColumnName]['Min']<descriptive[ColumnName]['Lesser']):
        lesser.append(ColumnName)
    if(descriptive[ColumnName]['Max']>descriptive[ColumnName]['Greater']):
        greater.append(ColumnName)
```

```python
lesser
```

```python
greater
```

```python
#Replacing the outliers in the column Transaction_Amount

dataset['Transaction_Amount'][dataset['Transaction_Amount']>descriptive['Transaction_Amount']['Greater']]=descriptive['Transaction_Amount']['Greater']
```
