```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from datetime import datetime
now=datetime.now()
```

```python
#Loading the dataset

dataset=pd.read_csv("Fraud_Prediction_preprocessed.csv",index_col=None)
dataset
```

```python
dataset.info()
```

```python
# Converting the categorical columns

col=dataset.select_dtypes(include=['object']).columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for x in col:
    dataset[x]=le.fit_transform(dataset[x])
```

```python
dataset
```

```python
# Splitting input and output data

indep=dataset.drop('Fraud_Label',axis=1)
dep=dataset['Fraud_Label']
```

```python
# Feature selection using SelectKBest Algorithm

def SelectKBest(indep,dep,n):
    from sklearn.feature_selection import SelectKBest
    from sklearn.feature_selection import chi2
    test=SelectKBest(score_func=chi2,k=n)
    fit=test.fit(indep,dep)
    selectk_features=fit.transform(indep)
    feature_names=indep.columns[fit.get_support(indices=True)].tolist()
    return selectk_features,feature_names
```

```python
kbest,Feature_Names=SelectKBest(indep,dep,3)
```

```python
kbest,Feature_Names
```

```python
model_performance=[]
```

```python
# Importing all the models from the classifier python file

from FS_Classifier import Split_scalar,cm_prediction,logistic,naive_bayes,KNN,randomforestclassifier,decisiontreeclassifier
```

```python
# Splitting the dataset into training and testing sets.
# Standardising the input data with Standard Scaler

X_train,X_test,y_train,y_test,sc=Split_scalar(kbest,dep)

# Training the models and evaluating the metrics for best model selection

classifier,accuracy,precision,recall,f1,score,cm,report=logistic(X_train,y_train,X_test,y_test)
model_performance.append({'Model': 'logistic', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC-ROC': score})
            
classifier,accuracy,precision,recall,f1,score,cm,report=KNN(X_train,y_train,X_test,y_test)  
model_performance.append({'Model': 'KNN', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC-ROC': score})
        
classifier,accuracy,precision,recall,f1,score,cm,report=naive_bayes(X_train,y_train,X_test,y_test)  
model_performance.append({'Model': 'Naive Bayes', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC-ROC': score})
        
classifier,accuracy,precision,recall,f1,score,cm,report=decisiontreeclassifier(X_train,y_train,X_test,y_test)  
model_performance.append({'Model': 'DecisionTree Classifier', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC-ROC': score})
        
classifier,accuracy,precision,recall,f1,score,cm,report=randomforestclassifier(X_train,y_train,X_test,y_test)  
model_performance.append({'Model': 'Random Forest Classifier', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1, 'AUC-ROC': score})
    
```

```python
Result = pd.DataFrame(model_performance)
```

```python
Result
```

# creating the model with the best performed one.
# From the above table, we can create the model with Random Forest Classifier since all the parameters are equal to 1 which denotes a very good performance of the model.

```python
from sklearn.ensemble import RandomForestClassifier
Classifier = RandomForestClassifier(criterion='gini', max_features='sqrt', n_estimators=100)
Classifier.fit(X_train,y_train)
y_pred=Classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
cm=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)
score=roc_auc_score(y_test,Classifier.predict_proba(X_test)[:,1])
print("Confusion Matrix\n",cm)
print("Classification report\n",report)
print("ROC_AUC score is ",score)
```

```python
# Features selected for model creation 

feature_scores = pd.DataFrame({'Feature_name':Feature_Names,'Score':Classifier.feature_importances_})
feature_scores=feature_scores.sort_values(by='Score',ascending=False)
feature_scores
```
